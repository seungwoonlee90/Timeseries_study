```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm
from statsmodels.tsa.filters.hp_filter import hpfilter

sns.set_context('talk')
```

```python
df = pd.read_csv('Data/macrodata.csv', index_col=0, parse_dates=True)
```

```python
df.head()
```

```python
fig, ax = plt.subplots(figsize=(20,8))

ax.plot(df.index, df.realgdp)
ax.spines[["top", "right", "left"]].set_visible(False)
plt.show()
```

```python
gdp_cycle, gdp_trend = hpfilter(df.realgdp, lamb=1600)
```

```python
df['trend'] = gdp_trend
```

```python
fig, ax = plt.subplots(figsize=(20,8))

ax.plot(df.index, df.realgdp, label='realgdp')
ax.plot(df.index, df.trend, label='trend')
ax.spines[["top", "right", "left"]].set_visible(False)
ax.legend()
plt.show()
```

```python
df_2007 = df['2007-01-01':]
```

```python
fig, ax = plt.subplots(figsize=(20,8))

ax.plot(df_2007.index, df_2007.realgdp, label='realgdp')
ax.plot(df_2007.index, df_2007.trend, label='trend')
ax.spines[["top", "right", "left"]].set_visible(False)
ax.legend()
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm

sns.set_context('talk')
```

```python
airline = pd.read_csv('Data/airline_passengers.csv', index_col='Month', parse_dates=True)
```

```python
airline
```

```python
airline = airline.dropna()
```

```python
fig, ax = plt.subplots(figsize=(20,8))

ax.plot(airline.index, airline['Thousands of Passengers'])
ax.spines[["top", "right", "left"]].set_visible(False)
plt.show()
```

```python
from statsmodels.tsa.seasonal import seasonal_decompose
```

```python
result = seasonal_decompose(airline['Thousands of Passengers'], model='multiplicative')
```

```python
trend = result.trend.reset_index()
seasonal = result.seasonal.reset_index()
resid = result.resid.reset_index()
```

```python
fig, ax = plt.subplots(figsize=(20,8))

ax.plot(trend.Month, trend.trend)
ax.spines[["top", "right", "left"]].set_visible(False)
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(20,8))

ax.plot(seasonal.Month, seasonal.seasonal)
ax.spines[["top", "right", "left"]].set_visible(False)
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(20,8))

ax.plot(resid.Month, resid.resid)
ax.spines[["top", "right", "left"]].set_visible(False)
plt.show()
```
